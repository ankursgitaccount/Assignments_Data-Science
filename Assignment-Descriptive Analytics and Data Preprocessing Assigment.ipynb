```python
import pandas as pd
df= pd.read_csv('sales_data_with_discounts.csv')
df
```

```python
#Descriptive Analytics for Numerical Columns
# We will calculate all the satistcal measure of numerical column

import numpy as np
from scipy import stats as st

# Statistical Measures of Avg Price
print("Statistical Measures of Avg Price:")
print("Mean:", np.mean(df['Avg Price']).round(3))
print("Mode:", st.mode(df['Avg Price']))
print("Median:", np.median(df['Avg Price']).round(3))
print("Variance:", np.var(df['Avg Price']).round(3))
print("Standard Deviation:", np.std(df['Avg Price']).round(3))
print("------"*25)

# Statistical Measures of Total Sales Value
print("\nStatistical Measures of Total Sales Value:")
print("Mean:", np.mean(df['Total Sales Value']).round(3))
print("Mode:", st.mode(df['Total Sales Value']))
print("Median:", np.median(df['Total Sales Value']).round(3))
print("Variance:", np.var(df['Total Sales Value']).round(3))
print("Standard Deviation:", np.std(df['Total Sales Value']).round(3))
print("------"*25)

# Statistical Measures of Discount Rate (%)
print("\nStatistical Measures of Discount Rate (%):")
print("Mean:", np.mean(df['Discount Rate (%)']).round(3))
print("Mode:", st.mode(df['Discount Rate (%)']))
print("Median:", np.median(df['Discount Rate (%)']).round(3))
print("Variance:", np.var(df['Discount Rate (%)']).round(3))
print("Standard Deviation:", np.std(df['Discount Rate (%)']).round(3))
print("------"*25)

# Statistical Measures of Discount Amount
print("\nStatistical Measures of Discount Amount:")
print("Mean:", np.mean(df['Discount Amount']).round(3))
print("Mode:", st.mode(df['Discount Amount']))
print("Median:", np.median(df['Discount Amount']).round(3))
print("Variance:", np.var(df['Discount Amount']).round(3))
print("Standard Deviation:", np.std(df['Discount Amount']).round(3))
print("------"*25)

# Statistical Measures of Net Sales Value
print("\nStatistical Measures of Net Sales Value:")
print("Mean:", np.mean(df['Net Sales Value']).round(3))
print("Mode:", st.mode(df['Net Sales Value']))
print("Median:", np.median(df['Net Sales Value']).round(3))
print("Variance:", np.var(df['Net Sales Value']).round(3))
print("Standard Deviation:", np.std(df['Net Sales Value']).round(3))
```

```python
#Data Visualisation
import matplotlib.pyplot as plt
plt.boxplot(df["Total Sales Value"])
plt.title("Boxplot of Total Sales Value")
plt.show()
```

```python
import matplotlib.pyplot as plt

# histogram for discount amount by BU
df["Discount Amount"].plot(kind="hist")
plt.title("Histogram of Discount Amount")
plt.xlabel( "Discount Amount")
plt.ylabel("BU")
plt.show()
```

```python
# Bar graph for Discount amount by Brand
df.groupby('Brand')["Discount Amount"].mean().plot(kind="bar")
plt.title("Average Discount Amount by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Discount Amount")
plt.show()
```

```python
#Bar graph for brand and Net Sales Value
import matplotlib.pyplot as plt
df.groupby('Brand')["Net Sales Value"].mean().plot(kind="bar")
plt.title("Net Sales Value of Brand")
plt.xlabel("Brand")
plt.ylabel("Net Sales Value")
plt.show()
```

```python
#Bar graph for Days and Net Sales Value
import matplotlib.pyplot as plt
df.groupby('Day')["Net Sales Value"].mean().plot(kind="bar")
plt.title("Net Sales Value on for each Day")
plt.xlabel("Day")
plt.ylabel("Net Sales Value")
plt.show()
```
